# Пример восстановления исходного изображения размером 200х200 пикселей по первым десяти ГК, если известны матрица счётов и матрица весов.

Для восстановления исходного изображения воспользуемся матрицами счётов и весов. 

```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
score_url = 'https://cdn.openedu.ru/ESZCBLZ/courseware/v1/ee3345b3500b7571a589d3caa49ec743/asset-v1:ITMOUniversity+ADVML+spring_2021_ITMO+type@asset+block/X_reduced_417.csv'
load_url = 'https://cdn.openedu.ru/ESZCBLZ/courseware/v1/ad22c96aa75c298f81ec21de6f334865/asset-v1:ITMOUniversity+ADVML+spring_2021_ITMO+type@asset+block/X_loadings_417.csv'
scores = pd.read_csv(score_url, header = None, sep = ';', dtype = float)
loadings = pd.read_csv(load_url, header = None, sep = ';', dtype = float)
```

Найдем матрицу восстановленных данных как произведение матрицы счётов на транспонированную матрицу весов. 

```python
values = np.dot(scores,loadings.T)
```

Создадим изображение на основе полученных значений при помощи библиотеки matplotlib

```python
plt.imshow(values, cmap='Greys_r')
```
